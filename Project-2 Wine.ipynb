```python
df1.drop('Color intensity', axis=1, inplace=True)
df1.head()
```

```python
x=df1.drop('Class',axis=1)
y=df1['Class']
print(x)
print(y)
```

```python
x.shape
```

```python
y.shape
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sx=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
best_rstate=0
accu=0
for i in range(30,200):
    train_x,test_x,train_y,test_y=train_test_split(sx,y,test_size=0.25,random_state=i)
    svc=SVC()
    svc.fit(train_x,train_y)
    pred=svc.predict(test_x)
    temp_acc=accuracy_score(test_y,pred)
    if temp_acc>accu:
        accu=temp_acc
        best_rstate=i

print(f'Best accuracy {accu*100} found on random_state {best_rstate}')
```

```python
train_x,test_x,train_y,test_y=train_test_split(sx,y,test_size=0.25,random_state=32)
svc=SVC()
svc.fit(train_x,train_y)
predsvc=svc.predict(test_x)
print(accuracy_score(test_y,predsvc))
print(confusion_matrix(test_y,predsvc))
print(classification_report(test_y,predsvc))
```

```python
mb=MultinomialNB()
mb.fit(train_x,train_y)
predmb=mb.predict(test_x)
print(accuracy_score(test_y,predmb))
print(confusion_matrix(test_y,predmb))
print(classification_report(test_y,predmb))
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(train_x,train_y)
preddtc=dtc.predict(test_x)
print(accuracy_score(test_y,preddtc))
print(confusion_matrix(test_y,preddtc))
print(classification_report(test_y,preddtc))
```

```python
svc1=SVC(kernel='poly')
svc1.fit(train_x,train_y)
predsvc1=svc1.predict(test_x)
print(accuracy_score(test_y,predsvc1))
print(confusion_matrix(test_y,predsvc1))
print(classification_report(test_y,predsvc1))
```

```python
svc2=SVC(kernel='linear')
svc2.fit(train_x,train_y)
predsvc2=svc2.predict(test_x)
print(accuracy_score(test_y,predsvc2))
print(confusion_matrix(test_y,predsvc2))
print(classification_report(test_y,predsvc2))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier()
ad.fit(train_x,train_y)
predad=ad.predict(test_x)
print(accuracy_score(test_y,predad))
print(confusion_matrix(test_y,predad))
print(classification_report(test_y,predad))
```

```python
ad1=AdaBoostClassifier(base_estimator=svc2, algorithm='SAMME')
ad1.fit(train_x,train_y)
predad1=ad1.predict(test_x)
print(accuracy_score(test_y,predad1))
print(confusion_matrix(test_y,predad1))
print(classification_report(test_y,predad1))
```

```python
ad2=AdaBoostClassifier(base_estimator=svc, algorithm='SAMME')
ad2.fit(train_x,train_y)
predad2=ad2.predict(test_x)
print(accuracy_score(test_y,predad2))
print(confusion_matrix(test_y,predad2))
print(classification_report(test_y,predad2))
```
