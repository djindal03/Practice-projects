```python
#Encoding all the columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l=list(range(1,24))
l1=[str(i) for i in l]
for v in l1:
    df[v]=le.fit_transform(df[v])
df
```

```python
#dropping columns with very low correlation with target variable
df.drop(['2','4','6','17'],axis=1,inplace=True)
```

```python
df.shape
```

```python
x=df.drop(['1'],axis=1)
y=df['1']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
best_rstate=0
accu=0
for i in range(30,200):
    train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=i)
    svc=SVC()
    svc.fit(train_x,train_y)
    pred=svc.predict(test_x)
    temp_acc=accuracy_score(test_y,pred)
    if temp_acc>accu:
        accu=temp_acc
        best_rstate=i

print(f'Best accuracy {accu*100} found on random_state {best_rstate}')
```

```python
from sklearn.model_selection import cross_val_score
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=151)
svc=SVC()
svc.fit(train_x,train_y)
predsvc=svc.predict(test_x)
svcscore=cross_val_score(svc, x, y, cv=10)
print('accuracy', accuracy_score(test_y,predsvc))
print('cv_score', svcscore.mean())
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(train_x,train_y)
preddtc=dtc.predict(test_x)
dtcscore=cross_val_score(dtc, x, y, cv=10)
print('accuracy', accuracy_score(test_y,preddtc))
print('cv_score', dtcscore.mean())
```

```python
mb=MultinomialNB()
mb.fit(train_x,train_y)
predmb=mb.predict(test_x)
mbscore=cross_val_score(mb, x, y, cv=10)
print('accuracy', accuracy_score(test_y,predmb))
print('cv_score', mbscore.mean())
```

```python
kn=KNeighborsClassifier()
kn.fit(train_x,train_y)
predkn=kn.predict(test_x)
knscore=cross_val_score(kn, x, y, cv=10)
print('accuracy', accuracy_score(test_y,predkn))
print('cv_score', knscore.mean())
```

```python
rf=RandomForestClassifier()
rf.fit(train_x,train_y)
predrf=rf.predict(test_x)
rfscore=cross_val_score(rf, x, y, cv=10)
print('accuracy', accuracy_score(test_y,predrf))
print('cv_score', rfscore.mean())
```
